```python
n = len(problemsList)
errors_list = []
idx = 2337

import json
import time

with open('errors.txt', 'a', encoding='utf-8') as errorfile:
    while idx < n:
        print(f"Starting : {idx}  |  ", end="")
        if idx % 50 == 0:
            print("In sleep mode... for 120s")
            time.sleep(120)
                            
        problem = problemsList[idx]
        title_slug = problem["titleSlug"]
        idx = int(problem["frontendQuestionId"])
        is_correct, data = getProblemStatement(title_slug)

        if is_correct:
            file_name = f"{'{:04d}'.format(idx)}_{title_slug}.json"
            json_data = json.dumps(data, indent=4)
            write_content_to_file(json_data, "Problems_Data/" + file_name)
        else:
            errors_list.append((idx, title_slug))
            print(data)
            errorfile.write(str(idx) + ",\n")
        
        print(f"Processed {idx} / {n}...")
```
